```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("Cornell-University/arxiv")

print("Path to dataset files:", path)
```

```python
import numpy as np
import pandas as pd
import os


rng = np.random.default_rng(42)

file_path = os.path.join(
    path,
    "arxiv-metadata-oai-snapshot.json"
)

sample_size = 1_000_000
chunk_size = 100_000

sample = pd.DataFrame()

for chunk in pd.read_json(
    file_path,
    lines=True,
    chunksize=chunk_size
):
    combined = pd.concat([sample, chunk], ignore_index=True)

    if len(combined) > sample_size:
        idx = rng.choice(
            len(combined),
            size=sample_size,
            replace=False
        )
        sample = combined.iloc[idx]
    else:
        sample = combined

df = sample.reset_index(drop=True)
```

```python
from google.colab import files

# save sample
output_file = "arxiv_random_sample.parquet"
df["id"] = df["id"].astype("string[pyarrow]")

df.to_parquet(
    output_file,
    index=False
)

print(f"Saved {len(df)} rows to {output_file}")

# download to your computer
files.download(output_file)
```
